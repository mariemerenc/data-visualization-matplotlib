```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading previous data
seattle_weather = pd.read_csv('seattle_weather.csv')
austin_weather = pd.read_csv('austin_weather.csv')

# Months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Adding month column to seattle_weather
seattle_weather['MONTH'] = seattle_weather['DATE'].apply(lambda x: months[(x - 1) % 12])

# Partitioning dataset
seattle_weather = seattle_weather[seattle_weather['STATION'] == 'USW00094290']

seattle_weather = seattle_weather.reset_index(drop=True)

# Setting index to date
seattle_weather.index = seattle_weather['DATE']

# Adding month column to austin_weather
austin_weather['MONTH'] = austin_weather['DATE'].apply(lambda x: months[(x - 1) % 12])
```

```python
# Changing plot style
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
plt.show()
```

```python
# Choosing a style
plt.style.use("ggplot")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
plt.show()
```

```python
# The bmh style

plt.style.use("bmh")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
plt.show()
```

### Guidelines for choosing plotting style

* Dark backgrounds are usually less visible;

* If color is important, consider choosing colorblind-friendly options;

* If you think that someone will want to print your figure, useless ink;

* If it will be printed in black-and-white, use the "grayscale" style.

```python
# Loading previous data

medals = pd.read_csv('medals_by_country_2016.csv', index_col=0)
```

```python
# Saving the figure as a PNG file

fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
fig.savefig("gold_medals.png")
```

```python
# Different file formats

fig.savefig("gold_medals.jpg")
fig.savefig("gold_medals.svg")

#Resolution
fig.savefig("gold_medals.png", dpi=300)
```
